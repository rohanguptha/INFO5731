```
import pickle
import numpy as np 
import pandas as pd 
```

```
df = pd.read_pickle('time_converted_df.pickle')
df.head()
```

```
df.LAT = df.LAT.astype(float)
df.LON = df.LON.astype(float)
```

```
df = df[(df['EVENT_TYPE'] == 'accident') | (df['EVENT_TYPE'] == 'minor accident')].head(1000)
```

```
df.head()
```

```
import folium
```

```
latitude = 32.7767
longitude = -96.7970
```

```
DFW_Map = folium.Map(
                location = [latitude, longitude], 
                zoom_start = 10)
```

```
incidents_accident = folium.map.FeatureGroup()
latitudes = list(df.LAT)
longitudes = list(df.LON)
labels = list(df.EVENT_TYPE)

for lat, lng, label in zip(latitudes, longitudes, labels):
  if label == 'accident':
    folium.Marker(
      location = [lat, lng], 
      popup = label,
      icon = folium.Icon(color='red', icon='info-sign')
     ).add_to(DFW_Map) 
  else:
    folium.Marker(
      location = [lat, lng], 
      popup = label,
      icon = folium.Icon(color='green', icon='info-sign')
     ).add_to(DFW_Map)
    
    
# add incidents to map
DFW_Map.add_child(incidents_accident)
DFW_Map
```
