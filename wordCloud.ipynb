```
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
```

```
from wordcloud import WordCloud, STOPWORDS
```

```
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
```

```
text = input()
```

```
mask = np.array(Image.open(path.join(d, "img.JPG")))
```

```
stopwords = set(STOPWORDS)
stopwords.add("Nope")
```

```
wc = WordCloud(background_color = "white", max_words=2000, mask = mask, stopwords = stopwords, contour_width=3, contour_color='white')
```

```
wc.generate(text)
```

```
wc.to_file(path.join(d, "text.png"))
```

```
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```
