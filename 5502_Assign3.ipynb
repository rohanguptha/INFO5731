```
import numpy as np
import matplotlib.pyplot as plt
```

```
pi_values = []
#circle origin (0.5, 0.5)
circle_x = 0.5
circle_y = 0.5
#circle radius
radius = 0.5
```

```
def Simulating(N):
  np.random.seed(19680801)
  x = np.random.rand(N)
  y = np.random.rand(N)
  g, ax = plt.subplots(figsize=(20,20))
  ax.set(xlim=(0, 1), ylim = (0, 1))
  #checking the point location
  inside_x = []
  outside_x = []
  inside_y = []
  outside_y = []
  for point in range(len(x)):
    if((circle_x - x[point]) * (circle_x - x[point]) + (circle_y - y[point]) * (circle_y - y[point]) <= radius * radius): 
      inside_x.append(x[point])
      inside_y.append(y[point])
    else:
      outside_x.append(x[point])
      outside_y.append(y[point])
  # Monte Carlo method
  pi = 4 * (len(inside_x) / N)
  pi_values.append(pi) 
  plt.plot(inside_x, inside_y, '.', color='red');
  plt.plot(outside_x, outside_y, '.', color='blue');
  a_circle = plt.Circle((circle_x, circle_y), radius, fill=None)
  ax.add_artist(a_circle)
  
```

```
simulate = [10, 500, 1000, 5000, 10000, 50000, 80000, 100000, 200000]
for N in range(len(simulate)):
  Simulating(simulate[N])
```
